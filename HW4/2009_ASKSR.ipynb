
Since kurtosis for most of the funds are negative, we cannot commpute alpha and cannot continue with equation 38. 
Instead we will revert back and use equation 20 


In equation 20, sharpe ratio is only adjusted for skewness. As a result, we can compare funds that have a return distribution with negative kutosis


```python
kurtosis = monthly.kurtosis()
skew = monthly.skew()
sigma = monthly.std()
mu = monthly.mean()

SR = mu/sigma
# Assuming CARA utiltiy 
beta = 1 

ASSR = SR*(1+ beta*skew/3*SR)**0.5
```

```python
monthly_ASSR = pd.DataFrame(ASSR.sort_values(ascending = False))
monthly_ASSR.reset_index(inplace = True)
monthly_ASSR.columns = ['Fund', 'Monthly ASSR']
monthly_ASSR['Rank'] = [x+1 for x in range(len(monthly_ASSR))]
monthly_ASSR
```

```python
kurtosis = weekly.kurtosis()
skew = weekly.skew()
sigma = weekly.std()
mu = weekly.mean()

SR = mu/sigma
# Assuming CARA utiltiy 
beta = 1 

ASSR = SR*(1+ beta*skew/3*SR)**0.5
```

```python
weekly_ASSR = pd.DataFrame(ASSR.sort_values(ascending = False))
weekly_ASSR.reset_index(inplace = True)
weekly_ASSR.columns = ['Name', 'Weekly ASSR']
# weekly_ASSR.set_index(['Fund'], inplace = True)
weekly_ASSR['Rank'] = [x+1 for x in range(len(weekly_ASSR))]
weekly_ASSR
```

```python
pd.concat([monthly_ASSR, weekly_ASSR], axis = 1)
```
