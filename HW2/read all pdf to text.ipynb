```python
import pandas as pd
df = pd.read_csv('/Users/Kang/Desktop/file_path.csv', encoding = 'big5hkscs')
df
```

```python
#process_pdf
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams

from io import BytesIO
from io import StringIO

from urllib.request import urlopen
from urllib.request import Request
from urllib.parse import quote 

def pdf_to_text(pdfname):

    # PDFMiner boilerplate
    rsrcmgr = PDFResourceManager()
    sio = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, sio, codec=codec, laparams=laparams)
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    
    # Extract text
    fp = open(pdfname, 'rb')
    for page in PDFPage.get_pages(fp):
        interpreter.process_page(page)
    fp.close()

    # Get text from StringIO
    text = sio.getvalue()

    # Cleanup
    device.close()
    sio.close()

    return text
```

```python
my_list = ['貳、基金投資範圍及投資特色', '參、投資本基金之主要風險', '肆、本基金適合之投資人屬性分析', '伍、基金運用狀況']


import re
perfect = 0
need_1_layer_check = 0
need_2_layer_check = 0
for i in range(len(df['0'])):
    path = df['0'][i]
    name = path[33:-4]
    results = pdf_to_text(path)
    
    start_end = []
    for j in my_list:
        counter = 0
        tem = re.search(j, results)
        if tem == None:
            counter += 1
            tem = re.search(j[2:], results)
            if tem == None:
                counter += 1
                tem = re.search(j[:2], results)
        
        if counter == 0:
            start_end.append(tem.start())
            start_end.append(tem.end())
        elif counter == 1:
            start_end.append(tem.start() - 2)
            start_end.append(tem.end())   
        else:
            start_end.append(tem.start())
            start_end.append(tem.end() + len(j[2:]))

    if counter == 1:
        need_1_layer_check += 1
    elif counter == 2:
        need_2_layer_check += 1
    else:
        perfect += 1
        print(i, 'done')
        
    with open('/Users/Kang/Desktop/output_folder/' + name + '.txt', 'a') as f:
        f.write(results[start_end[1]:start_end[2]])
        f.write(results[start_end[3]:start_end[4]])
        f.write(results[start_end[5]:start_end[6]])
        f.close()
```

```python
print(perfect, need_1_layer_check, need_2_layer_check)
```
